**BAR CHART FOR NUMBER OF PEOPLE UNEMPLOYMENT**

```python
fig=plt.figure() 
ax0=fig.add_subplot(1,2,1)
df[ :10].plot(kind='bar',color='blue', figsize=(30,5), ax=ax0)
ax0.set_title('top 10 States with highest')
ax0.set_xlabel('State')
ax0.set_ylabel('Number of people unemployment %')
```

```python
fig=plt.figure()
ax0=fig.add_subplot(1,2,1)
df[ :12].plot(kind='bar', color='green', figsize=(30,5), ax=ax0)
ax0.set_title('Month with Highest unemployment')
ax0.set_xlabel('Region')
ax0.set_ylabel('Number of people unemployment(in %)')
```

**BAR CHART FOR ESTIMATED EMPLOYED**

```python
df=df[['Region', 'Estimated Employed']].groupby('Region').sum().sort_values('Estimated Employed')
df.head(10)
fig=plt.figure()
ax1=fig.add_subplot(1,2,2)

df[ :10].plot(kind='bar', color='pink', figsize=(15,6), ax=ax1)
ax1.set_title('Estimated Employed people in each state')
ax1.set_title('State')
ax1.set_ylabel('Number of Estimated Employed')
```

**SCATTER PLOT FOR EMPLOYMENT**

```python
x, y, scale = np.random.randn(3, 100)
fig, ax = plt.subplots()


ax.scatter(x=x, y= y, c=scale, s=np.abs(scale)*500)
ax.set(title="Estimation of Employment")
plt.show()
```
